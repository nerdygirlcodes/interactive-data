```python
import numpy as np
import pandas as pd
from bokeh.models import HoverTool
from bokeh.plotting import figure, show, reset_output
from bokeh.io import output_notebook, output_file
```

```python
twitch = pd.read_csv('https://raw.githubusercontent.com/Hamidest/InteractiveData_EMAT/main/twitchdata-update.csv')
twitch
```

```python
tmp = twitch.sort_values('Stream time(minutes)', ascending=False)
tmp

# sort by language, find most amount of stream time --->
```

```python
twitch_group = twitch.groupby('Language')
counts = twitch_group.count()
```

```python
counts = counts.sort_values('Stream time(minutes)', ascending=False)
counts
```

```python
x = list(counts.index)
y = counts['Channel']
output_notebook()
output_file('in-class-figure.html')


fig = figure(x_range=x, sizing_mode="stretch_width")
fig.line(x, y)

# opens in new tab
show(fig)
```

```python
hover1 = HoverTool(
    tooltips=[("Value", "@x / @y")],
    mode='vline'
)

fig2 = figure(
    x_range=x,
    sizing_mode="stretch_width",
    height=300,
    title="Twitch Streamers by Spoken Language",
    x_axis_label="Spoken Language",
    y_axis_label="Number of Streamers",
    toolbar_location=None,
    tools=[hover1]
)

fig2.line(x, y, line_color="green", legend_label="Number of Streamers", line_width=2)

fig2.ygrid.grid_line_alpha = 0.8
fig2.ygrid.grid_line_dash = [6, 4]

show(fig2)
```
