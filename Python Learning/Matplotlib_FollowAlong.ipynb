## Intro to Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()  # create a figure containing a single axes
ax.plot([1,2,3,4], [1,4,2,3]);  # plot some data on the axes
```

## Types of inputs

```python
np.random.seed(19680801) # seed the random number generator
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

## Coding Styles  (implicit + explicit)

```python
# Object Oriented (explicit) style

x = np.linspace(0, 2, 100) # sample data

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained') 
# layout is not in the documentation, what is this doing??

ax.plot(x, x, label='linear') # plot some data on the axes
ax.plot(x, x**2, label='quadratic') # more data
ax.plot(x, x**3, label='cubic') # more data

ax.set_xlabel('x label') # add x-label
ax.set_ylabel('y label') # add y-label
ax.set_title('simple OO plot') # add title
ax.legend() # add legend
```

```python
# pyplot style

x = np.linspace(0, 2, 100) # sample data

plt.figure(figsize=(5, 2.7), layout='constrained')

plt.plot(x, x, label='linear') # plot on (implicit) axes
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('simple pyplot plot')
plt.legend()
```

## Making a helper function

```python
def my_plotter(ax, data1, data2, param_dict):
    # helper function to make a graph
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100) 
#make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
