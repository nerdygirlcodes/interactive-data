```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot
```

```python
cces = pd.read_csv('cces_example_data.csv')
```

```python
cces = cces.astype({
    'sex': 'category',
    'pid3': 'category',
    'Evaluation': 'category'
})
cces
```

```python
# '*' defines an 'interaction term'
rt_model = ols('Real_Threat ~ pid3 + ConMediaUse + pid3*ConMediaUse', data = cces).fit()
rt_model.summary()
```

```python
sns.regplot(data = cces, y = 'Real_Threat', x = 'ConMediaUse', x_jitter = .08, y_jitter = .08)
```

```python
sns.jointplot(data = cces, y = 'Real_Threat', x = 'ConMediaUse', x_jitter = 0.08, y_jitter = 0.08, kind = 'reg')
```

```python
sns.lmplot(data = cces, y = 'Real_Threat', x = 'ConMediaUse', x_jitter = 0.0, y_jitter = 0.0, hue = 'pid3')
```

```python
sns.lmplot(data = cces, y = 'Real_Threat', x = 'LibMediaUse', x_jitter = 0.0, y_jitter = 0.0, hue = 'pid3'))
```

```python
sns.lmplot(
    data = cces, 
    y = 'Real_Threat', 
    x = 'ConMediaUse', 
    x_jitter = 0.08, 
    y_jitter = 0.08, 
    hue = 'pid3', 
    scatter_kws = {'alpha': 0.05}
)
```

```python
sns.jointplot(data = cces, y = 'Real_Threat', x = 'ConMediaUse', hue = 'pid3')
```
