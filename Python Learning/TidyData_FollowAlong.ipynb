```python
df_billboard = pd.read_csv('./raw/billboard.csv', encoding='mac_latin2')
df_billboard.head(10)
```

```python
# MELTING!! ----->
id_vars = ['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered', 'date.peaked']
df_billboard = pd.melt(frame=df_billboard, id_vars=id_vars, var_name="week", value_name="rank")

# FORMATTING!! ----->
df_billboard['week'] = df_billboard['week'].str.extract('(\d+)', expand=False).astype(int)
df_billboard['rank'] = df_billboard['rank'].astype(int)

# CLEAN UNNECESSARY ROWS ----->
df_billboard = df_billboard.dropna()

# CREATE "DATE" COLUMNS ----->
df_billboard['date'] = df_billboard.to_datetime(df_billboard['date.entered']) + df_billboard.to_timedelta(df_billboard['week'], unit='w') - df_billboard.DateOffset(weeks=1)
df_billboard = df_billboard[['year', 'artist.inverted', 'track', 'time', 'genre', 'week', 'rank', 'date']]
df_billboard = df_billboard.sort_values(ascending=True, by=['year', 'artist.inverted', 'track', 'week', 'rank'])

df_billboard.head(10)
```
