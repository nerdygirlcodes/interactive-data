```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
commute = pd.read_csv('commute.csv', header = 0)
```

```python
# create a new figure
fig_1, ax_1 = plt.subplots()

# sort the data by vehicle
veh_group = commute.groupby('Vehicle')

# create a list of heights
heights = [
    veh_group['CommuteTime'].max()['Passat'],
    veh_group['CommuteTime'].min()['Passat'],
    veh_group['CommuteTime'].max()['Street Triple'],
    veh_group['CommuteTime'].min()['Street Triple']
]

# make a horizontal bar plot by adding 'h' to make 'barh'
ax_1.barh([0, 1, 3, 4], heights)

# make the y-axis reflect what we're showing
ax_1.set_yticks([0.5, 3.5], ['Passat', 'Street Triple'])

# set the figure title
ax_1.set_title('Commute Time in Minutes by Vehicle')

# create labels for street triple
ax_1.text(15, 4, 'sup')
ax_1.text(18, 3, heights[2])

# create labels for passat
ax_1.text(17, 1, heights[1])
ax_1.text(20, 0, heights[0])

# show the figure
plt.show();
```

```python
# create a new figure with 2 axes
fig_2, ax_2 = plt.subplots(1, 2)

# sort the data by vehicle
sort_group = commute.groupby(['Vehicle','Weather']).count()['TripID']
sort_group

# create a pie for each vehicle with labeling
ax_2[0].pie(sort_group['Passat'], labels=sort_group['Passat'].index, colors=['blue','gray','orange'])
ax_2[0].set_xlabel('Passat')
ax_2[1].pie(sort_group['Street Triple'], labels=sort_group['Street Triple'].index, colors=['blue','orange'])
ax_2[1].set_xlabel('Street Triple')

# show the figures
plt.show()
```
