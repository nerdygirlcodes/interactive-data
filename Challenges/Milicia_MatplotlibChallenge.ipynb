#### Matplotlib Categorical Heatmap
We use heatmaps similar to this style quite often in my field.  The type of information we'd need would be two different verticies which cross reference each other at variable values / amounts.  (e.g., store location x service type = amount of appts for each) 

Demo code copied from:

https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html#sphx-glr-gallery-images-contours-and-fields-image-annotated-heatmap-py 

Data altered to fit different example data, however, rendering code was copied 1:1.

```python
# altered demo data, but code copied from above (as noted)
services = ["roll", "tires", "oil change", "break check",
              "alignment", "consultation", "other"]
locations = ["Vans Tires", "Conrads", "Goodyear Auto Service",
           "Just Tires", "TSN", "Partnerships (AAA)", "Dealer"]

appointments = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 7.4, 1.0, 2.7, 0.0, 4.0],
                    [1.1, 2.4, 5.8, 4.3, 1.9, 4.4, 1.7],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(appointments)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(locations)), labels=locations)
ax.set_yticks(np.arange(len(services)), labels=services)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(services)):
    for j in range(len(locations)):
        text = ax.text(j, i, appointments[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Service Appointments (Appts/Day over 1 Month)")
fig.tight_layout()

plt.show()
```

## Task #3

*Download cces_example_data.csv using pandas.read_csv()*
- *This data already follows tidy data principles*
- *The variables `sex`, `pid3`, `nonwhite`, and `Evaluation` are all categorical variables, the rest can be treated as continuous*

```python
# read the csv data and add to DataFrame
cces = pd.read_csv('raw/cces_example_data.csv', header = 0)

# set columns as categorical
cces_clean = cces.astype({
    'sex':'category',
    'pid3':'category',
    'nonwhite':'category',
    'Evaluation':'category'
})
```

## Task #4

*Create and show in your notebook an empty figure that contains two axes*
- *Set the title for the first axis as "Age Histogram"*
- *Set the title for the second axis as "Political Identity"*

```python
# create a new figure with two axes - 1 row, 2 axes
fig_1, ax_1 = plt.subplots(1, 2, figsize=(5, 3), layout='constrained')

# name each axes
ax_1[0].set_title('Age Histogram')
ax_1[1].set_title('Political Identity')

# show figures
plt.show()
```

## Task #5

*In a new code cell, create and show a figure that includes one axis with a bar plot of the `Evaluation` variable*

```python
# create a new figure
fig_2, ax_2 = plt.subplots(figsize=(5, 3), layout='constrained')

# create a var to group data by the evaluation column; only grab the 'caseid' values (height)
eval_count = cces_clean.groupby('Evaluation').count()['caseid']

# create a var for the evaluation categories (x)
eval_cats = eval_count.index.categories

# draw each bar with the grouped data
ax_2.bar(eval_cats, eval_count)

# rename the x-axis values
ax_2.set_xticks(eval_cats, ['Legitimate', 'Unsure', 'Illigitimate', 'Other'])

# show figure    
plt.show()
```

## Task #6

*In a new code cell, create and show a figure that includes one axis with a scatter plot of the `ConMediaUse` and `ImmiPhobia` variable.*

```python
# create a new figure 
fig_3, ax_3 = plt.subplots(figsize=(5, 3), layout='constrained')

# create a scatter plot (x, y) + bonus: add alpha
ax_3.scatter(cces_clean['ConMediaUse'], cces_clean['ImmiPhobia'], alpha = .05)

# name the axis
ax_3.set_xlabel('ConMediaUse')
ax_3.set_ylabel('ImmiPhobia')

# show the figure
plt.show()
```
